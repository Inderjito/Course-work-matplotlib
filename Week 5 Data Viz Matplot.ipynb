```python
import numpy as np

fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)

ax1.hist(np.random.standard_normal(100), bins=20, color="black",alpha =0.3)

# create an array of 30 evenly space value between 0-29 ( represent x- axis)
# np.arange(30)+3 * np.random.standard_normal(30))  generate 30 y-values by
# adding a random normal distribution with 0 mean and STD 1*3 each value in array (30)
# result in 3 slope line

ax2. scatter(np.arange(30),np.arange(30)+3 * np.random.standard_normal(30));


# Create array of 50 values from STD normal  distribution with mean 0 & S normal D ) and Standard deviation 1
# cumsum() is a method in Numpy array that Calculate cumulative sum of value in array, used  as X- vlaue for line plot
ax3. plot(np.random.standard_normal(50).cumsum(),color="green",linestyle ="dashed")

```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)
ax1.plot(np.random.standard_normal(30).cumsum(), color="black", linestyle="dashed", marker="o")
ax2.plot(np.random.standard_normal(30).cumsum(), color="green", linestyle="dashed", marker="o")
ax3.plot(np.random.standard_normal(30).cumsum(), color="blue", linestyle="dashdot", marker="o")
ax4.plot(np.random.standard_normal(30).cumsum(), color="blue", linestyle="dashdot", marker="x")
```

```python
import numpy as np
fig = plt.figure()
data = np.random.standard_normal(30).cumsum()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)
ax1.plot(data, color="blue", linestyle="dashed")
ax2.plot(data, color="blue", linestyle="dashed")
ax2.plot(data, color="black", linestyle="dashed", drawstyle="steps-post")
ax3.plot(data, color="blue", linestyle="dashed")
ax3.plot(data, color="black", linestyle="dashed", drawstyle="steps-mid")
ax4.plot(data, color="blue", linestyle="dashed")
ax4.plot(data, color="black", linestyle="dashed", drawstyle="steps")
```

```python
import numpy as np
fig = plt.figure()
data1 = np.random.standard_normal(30).cumsum()
data2 = np.random.standard_normal(30).cumsum()
data3 = np.random.standard_normal(30).cumsum()
ax = fig.add_subplot(1, 1, 1)
ax.set_xticks([0, 10, 20, 30])
ax.set_xticklabels(["one", "two", "three", "four"])
ax.set(title="Annotated Plot", xlabel="stages", ylabel="values")
ax.plot(data1, color="black", label="one")
ax.plot(data2, color="green", label="two")
ax.plot(data3, color="red", label="three")
ax.legend()
```

```python
tips = pd.read_csv("tips.csv")
tips.head()
```

```python


party_counts = pd.crosstab(tips['day'],tips['size']) # use corss tab to create df on day
party_counts = party_counts.reindex(index =['Thur','Fri','Sat','Sun']) # name index thur, fri, etc
party_counts
```

```python
party_counts = party_counts.loc[:,2:5] #  slicing  from df 2:5 should be included. After , indict that 2,3,4,5 included in new df
party_pcts = party_counts.div(party_counts.sum(axis ='columns'), axis ='index') #  cal percentage by divide the vlaue in party_count in each row
party_pcts # contain result ( percentage vlaue of each party count)
```
